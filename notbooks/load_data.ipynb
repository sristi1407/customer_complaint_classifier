```python
# Install dependencies as needed:
# pip install kagglehub pandas

from pathlib import Path

import pandas as pd
import kagglehub
from kagglehub.exceptions import KaggleApiHTTPError


# Try one or more handles; update with the exact dataset slug from Kaggle.
DATASET_HANDLE_CANDIDATES = [
    "selener/consumer-complaint-database",
]

# File name to extract from the downloaded dataset folder (if present)
KAGGLE_FILE_NAME = "consumer_complaints.csv"

# Save final CSV here for easy reuse
project_root = Path.cwd().parent
raw_data_dir = project_root / "data" / "raw_data"
raw_data_dir.mkdir(parents=True, exist_ok=True)
output_csv = raw_data_dir / "consumer_complaints.csv"


def load_from_kaggle(handle: str, preferred_file: str) -> pd.DataFrame:
    """Download dataset, then load the preferred CSV (or first CSV found)."""
    dataset_dir = Path(kagglehub.dataset_download(handle))
    csv_files = list(dataset_dir.rglob("*.csv"))
    if not csv_files:
        raise FileNotFoundError(f"No CSV files found in downloaded dataset: {dataset_dir}")

    preferred_path = dataset_dir / preferred_file
    selected_csv = preferred_path if preferred_path.exists() else csv_files[0]
    print(f"Using CSV: {selected_csv}")
    return pd.read_csv(selected_csv)


last_error = None
df = None

for handle in DATASET_HANDLE_CANDIDATES:
    try:
        print(f"Trying Kaggle handle: {handle}")
        df = load_from_kaggle(handle, KAGGLE_FILE_NAME)
        print(f"Loaded dataset from: {handle}")
        break
    except KaggleApiHTTPError as e:
        last_error = e
        print(f"Kaggle API error for '{handle}': {e}")
    except Exception as e:
        last_error = e
        print(f"Failed for '{handle}': {e}")

if df is None:
    raise RuntimeError(
        "Could not download dataset from Kaggle. "
        "Verify the dataset handle and your Kaggle API credentials (~/.kaggle/kaggle.json)."
    ) from last_error

df.to_csv(output_csv, index=False)
print(f"Saved to: {output_csv}")
print(f"Shape: {df.shape}")
display(df.head())
```
